```python
# load ouput txt file
with open('training_record.txt','r') as f:
    lines = f.readlines()
```

```python
losses = []
accuracies = []
```

```python
for line in lines:
    if ',' in line:
        tl = line.split(',')
        lossText = tl[1]
        accText = tl[2].split('(')[0]
        loss = float( lossText.split('=')[1].strip() )
        acc =  float( accText.split('=')[1].strip() )
        losses.append(loss)
        accuracies.append(acc)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6), dpi=80)
plt.subplot(2,1,1)
plt.plot(losses)
plt.subplot(2,1,2)
plt.plot(accuracies)
plt.show
```
